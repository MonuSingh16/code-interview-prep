#### Statement
Given the head of a singly linked list, return the middle node of the linked list. If the number of nodes in the linked list is even, there will be two middle nodes, so return the second one.



```python

def get_middle_node(head):

    # Create two pointers, slow and fast ,initially pointing to the head
    slow, fast = head, head
    
    # Traverse the linked list until fast reaches at the last node or NULL
    while fast and fast.next:

        # Move slow pointer one step ahead
        slow = slow.next
        
        # Move fast pointer two steps ahead
        fast = fast.next.next
        
    # Return the slow pointer
    return slow
```

```python
# Template for traversing a linked list

def traverse_linked_list(head):
    current, nxt = head, None
    while current:
      nxt = current.next
      current = nxt
```

```python
# Template for reversing a linked list

def reverse_linked_list(head):
	prev, curr = None, head
	while curr:
		nxt = curr.next
		curr.next = prev
		prev = curr
		curr = nxt
	return prev
```
